```python
import pandas as pd
```

```python
data =pd.read_csv("housing.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
x1=data["longitude"]
```

```python
x2=data["latitude"]
```

```python
from sklearn.cluster import KMeans
```

```python
k=KMeans(n_clusters=2)
```

```python
k
```

```python
k.fit(x1.to_frame(),x2.to_frame())
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x=data['longitude'],y=data['latitude'],c=k.labels_)
```

```python
k.cluster_centers_
```

```python
x3=data['total_rooms']
```

```python
x4=data['total_bedrooms']
```

```python
from sklearn.cluster import KMeans
```

```python
k=KMeans(n_clusters=4)
```
